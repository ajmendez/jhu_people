```python
import json
import requests
from pprint import pprint
from bs4 import BeautifulSoup
```

```python
URL = 'http://physics-astronomy.jhu.edu/people/directory/'
soup = BeautifulSoup(requests.get(URL).text)
```

```python
s = soup.find('section', {'class':'content'})
h = ['name','position','phone','email','offie','category']
# for header in s.findAll('th'):
#     h.append(header.find('div').text)
# print h

out = []
for person in s.findAll('tr'):
    if 'Full Name' in person.prettify():
        continue
    tmp = {key:value.text for key,value in zip(h,person.findAll('td'))}
    tmp['category']=''
    if tmp['category'] in ['Misc','Lab'] or 'Room' in tmp['name']:
        continue
    out.append(tmp)
#     break
# pprint(out)
postdocs = [x for x in out 
            if x['position'] in ['Postdoctoral Fellow','Assistant Research Scientist']
            or x['category'] == 'Postdoctoral Fellow']
students = [x for x in out 
            if x['category'] in [u'Graduate Student'] or 
            x['position'] in [u'Graduate Assistant']]
faculty = [x for x in out if 'Professor' in x['position']]
# print len(students)
# pprint(students)
# print len(postdocs)
# pprint(postdocs)
```

```python
json.dump(faculty,open('faculty.json','w'))
json.dump(postdocs,open('postdocs.json','w'))
json.dump(students,open('students.json','w'))
```

```python
def build_link(people):
    # lastname_firstInit
    print 'N_people: {}'.format(len(people))
    names = []
    for x in people:
        tmp = x['name'].split(',')
        if ' ' in tmp[0]: tmp[0] = tmp[0].split()[0]
        names.append('{}_{}'.format(tmp[0],tmp[1].strip()[0]))
    ors = ['OR']*(len(names)-1)
    URL = 'http://arxiv.org/find/astro-ph/1/au:+{}+{}/0/1/0/2013,2014/0/1?per_page=100'
    return URL.format('+'.join(ors), '+'.join(names))
print "Faculty Link:"
print build_link(faculty)
```
